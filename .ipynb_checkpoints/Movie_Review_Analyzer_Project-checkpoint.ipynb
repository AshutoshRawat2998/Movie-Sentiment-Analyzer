```python
import numpy as np
import pandas as pd
import seaborn as sns
import nltk
import re
import string
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from gensim.models import Word2Vec
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
```

```python
df = pd.read_csv('IMDB Dataset.csv')
df.head()
```

```python
df.shape
```

## Text Cleaning / Preprocessing

```python
# Remove html tags
def remove_html(text):
    text = text.lower()
    text = re.sub(r'<[^>]*>', '', text)
    return text

# Remove stopwrods and punctuations
exclude = stopwords.words('english')
strp = string.punctuation
def remove_st_punc(text):
    text = text.translate(str.maketrans('', '', strp))
    word = [char for char in text.split() if char.lower() not in exclude]
    return ' '.join(word)
```

```python
df['review'] = df['review'].apply(remove_html)
```

```python
df['review'] = df['review'].apply(remove_st_punc)
```

```python
df['review'][5]
```

## Text Vectorization

```python
# Import Libararies
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features = 5000)
X = vectorizer.fit_transform(df['review'])
y = df['sentiment'].apply(lambda x: 1 if x == 'positive' else 0)
```

## Train-Test Split for TF-IDF Model

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

## Model Training (TF-IDF + Logistic Regression)

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

## Model Evaluation (TF-IDF Model)

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

## Word2Vec Tokenization

```python
# Assuming 'reviews' is a list or pandas series of text reviews
tokenized_reviews = [nltk.word_tokenize(review) for review in df['review']]
```

```python
# Training Word2Vec Embeddings
w2v_model = Word2Vec(sentences=tokenized_reviews, vector_size=100, window=5, min_count=2, workers=4)
```

## Averaging Word2Vec Vectors

```python
def vectorize_review(words, model, size=100):
    vec = np.zeros(size)
    count = 0
    for w in words:
        if w in model.wv:
            vec += model.wv[w]
            count += 1
    if count != 0:
        vec /= count
    return vec

X_w2v = np.array([vectorize_review(tokens, w2v_model) for tokens in tokenized_reviews])
```

## Random Forest with Word2Vec Vectors

```python
X_train, X_test, y_train, y_test = train_test_split(X_w2v, y, test_size=0.2, random_state=42)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
print("RF Accuracy:", rf.score(X_test, y_test))
```

## LSTM Model Preparation

```python
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(df['review'])
sequences = tokenizer.texts_to_sequences(df['review'])
X_seq = pad_sequences(sequences, maxlen=100)
```

```python
# Build LSTM model
model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=100, input_length=100))
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))
```

## LSTM Training and Validation

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_seq, y, epochs=5, batch_size=64, validation_split=0.2)
```

## LSTM Test Evaluation

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

## Add/Adjust Dropout for Overfitting

```python
max_len = max(len(seq) for seq in sequences)
X_seq = pad_sequences(sequences, maxlen=max_len)
```

```python
X_seq = np.expand_dims(X_seq, -1)  # Shape becomes (samples, sequence_length, 1)
num_features = X_seq.shape[2]      # This will be 1
max_len = X_seq.shape[1]
```

## LSTM Model with Dropout

```python
model = Sequential()
# LSTM with dropout on inputs (dropout) and recurrent connections (recurrent_dropout)
model.add(LSTM(128, input_shape=(max_len, num_features), dropout=0.3, recurrent_dropout=0.3))
# Optional additional dropout layer after LSTM
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))
```

## LSTM Training (Dropout Model)

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=20, batch_size=64, validation_split=0.2)
```
